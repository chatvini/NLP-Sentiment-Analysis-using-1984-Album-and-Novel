# Textual Data Exploration

```python
#Van Halen 1984 album

album_folder= "/Users/macmojave/Downloads/Album Van Halen 1984"
songs = os.listdir(album_folder)
songs
```

```python
album_corpus =[]

for lyrics in songs:
    with open(album_folder+'/'+lyrics, encoding='utf8') as f_input:
        album_corpus.append(f_input.read())
```

```python
#func to remove special characters
def remove_special_char(text):
    result = re.sub(r'[^a-zA-A0-9 \\\n]',"",text)
    return result
```

```python
album_corpus
```

```python
album_corpus1= list(map(str.lower, album_corpus))
album_corpus1=list(map(remove_special_char, album_corpus1))
```

```python
album_word_list= []

for lyrics in album_corpus1:
    for word in lyrics.split():
        album_word_list.append(word)
        
        
album_word_list
```

```python
len(album_word_list)
```

```python
album_word_count_unique = Counter(album_word_list)
len(album_word_count_unique)
```

```python
#1984 novel

novel= "/Users/macmojave/Downloads/1984-novel-George Orwell.txt"
novel_corpus =[]

with open(novel,encoding='utf8') as f_input:
    novel_corpus.append(f_input.read())
```

```python
novel_corpus
```

```python
novel_corpus1= list(map(str.lower, novel_corpus))
novel_corpus1=list(map(remove_special_char, novel_corpus1))
```

```python
novel_word_list = novel_corpus1[0].split()
```

```python
novel_word_list
```

```python
len(novel_word_list)
```

```python
len(Counter(novel_word_list))
```

# Word Clouds

```python
stopwords = list(get_stop_words('english'))
```

```python
stopwords
```

```python
#creating a custom list of stopwords
stop_list = ["intro","verse","chorus","prechorus"]
#Adding custom list of stopwords 

# entend()function is used to add custom stopwords 
stopwords.extend(stop_list)
print(stopwords)
```

```python
clean_album_word_list = []

for word in album_word_list:
    if word not in stopwords:
        clean_album_word_list.append(word)
```

```python
clean_album_word_list
```

```python
import nltk

freq = nltk.FreqDist(clean_album_word_list)
```

```python
freq
```

```python
freq.plot(50,cumulative=False)
```

```python
count_clean_album_word_list = Counter(clean_album_word_list)

wc_album = WordCloud(width=1000, height= 500).generate_from_frequencies(count_clean_album_word_list)

plt.figure(figsize=(20,10))
plt.figtext(0.5,0.9,"WordCloud of 1984 Album", fontsize=60,ha='center')
plt.imshow(wc_album)
plt.axis("off")
plt.show
```

```python
clean_novel_list =[]
for word in novel_word_list:
    if word not in stopwords:
        clean_novel_list.append(word)
```

```python
clean_novel_list
```

```python
freq1 = nltk.FreqDist(clean_novel_list)
```

```python
freq1.plot(50,cumulative=False)
```

```python
count_clean_novel_word_list = Counter(clean_novel_list)

wc_novel = WordCloud(width=1000, height= 500).generate_from_frequencies(count_clean_novel_word_list)

plt.figure(figsize=(20,10))
plt.figtext(0.5,0.9,"WordCloud of 1984 Novel", fontsize=60,ha='center')
plt.imshow(wc_novel)
plt.axis("off")
plt.show
```


# NLP Summarization

```python
nltk.download('punkt')

lsa = LsaSummarizer(Stemmer('english'))
lsa.stop_words = get_stop_words('english')
parser = PlaintextParser.from_string(' '.join(album_corpus), Tokenizer('english'))
lsa_summary = lsa(parser.document,10)
```

```python
lsa_summary
```

```python
lsa_summary_list = [str(sentence) for sentence in lsa_summary]

summary_album = " ".join(lsa_summary_list)
print("Summary of 1984 Album \n" + summary_album)
```

```python
parser = PlaintextParser.from_string(' '.join(novel_corpus), Tokenizer('english'))
lsa_summary = lsa(parser.document,10)

lsa_summary_list = [str(sentence) for sentence in lsa_summary]

summary_novel = " ".join(lsa_summary_list)
print("Summary of 1984 Novel \n" + summary_novel)
```
